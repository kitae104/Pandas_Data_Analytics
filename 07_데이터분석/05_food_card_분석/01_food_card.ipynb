```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv("1_foodcard.csv")

# Display the first few rows of the DataFrame to get an overview
df.head()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
```

```python


# Filter the dataframe for the year 2020
df_2020 = df[df['crym'].astype(str).str.startswith('2020')]

# Group by age group and food type and sum up the usage amount
grouped = df_2020.groupby(['agegrp_dc', 'tco_btc_nm'])['tco_btc_u_am'].sum().unstack()

# Plot the data
plt.figure(figsize=(12, 8))

for column in grouped.columns:
    plt.plot(grouped.index, grouped[column], label=column, marker='o')

plt.title('2020년 연령대별 사용금액 (음식 유형별)')
plt.xlabel('연령대')
plt.ylabel('사용금액')
plt.legend(loc='upper right')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
